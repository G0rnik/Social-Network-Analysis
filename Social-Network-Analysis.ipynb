```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the adjacency matrix based on the given table
nodes = ["nn", "kk", "ss", "mm", "rr", "zz", "yy", "pp", "oo"]
edges = [
   ("nn", "ss", 5), ("nn", "mm", 4), ("nn", "kk", 5), ("nn", "zz", 1), ("nn", "yy", 1),
   ("kk", "ss", 1), ("kk", "rr", 4), ("kk", "mm", 5), ("kk", "zz", 1), ("kk", "yy", 1),
   ("ss", "mm", 3), ("ss", "zz", 3), ("ss", "yy", 3), ("kk", "pp", 1), ("kk", "oo", 1),
   ("rr", "kk", 4), ("rr", "zz", 1), ("rr", "yy", 1), ("rr", "pp", 3), ("rr", "oo", 4),
]


# Create a weighted graph
G = nx.Graph()


# Add weighted edges
for src, dst, weight in edges:
   G.add_edge(src, dst, weight=weight)


# Compute inverse weights to make strong edges shorter and weak edges longer
inverse_weights = {edge: 1 / data["weight"] for edge, data in G.edges.items()}


# Compute layout using inverse weights
pos = nx.spring_layout(G, weight="weight", scale=2, k=2.5, seed=42)


# Draw the graph
plt.figure(figsize=(8, 8))
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=2000, font_size=10)


# Draw edge labels
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)


# Show plot
plt.title("Social Network Graph (Weak Links Stretched, Strong Links Shortened)")
plt.show()




# Calculate node strengths
node_strength = {node: sum(data['weight'] for _, _, data in G.edges(node, data=True)) for node in G.nodes()}


# Sort node strengths from highest to lowest
sorted_strengths = dict(sorted(node_strength.items(), key=lambda item: item[1], reverse=True))


# Print sorted strengths
print(f"Node Strengths (Descending): {sorted_strengths}")
```

```python
Gw = nx.Graph()

Gw.add_edge("a", "b", weight=0.6)
Gw.add_edge("a", "c", weight=0.2)
Gw.add_edge("c", "d", weight=0.1)
Gw.add_edge("c", "e", weight=0.7)
Gw.add_edge("c", "f", weight=0.9)
Gw.add_edge("a", "d", weight=0.3)

elarge = [(u, v) for (u, v, d) in Gw.edges(data=True) if d["weight"] > 0.5]
esmall = [(u, v) for (u, v, d) in Gw.edges(data=True) if d["weight"] <= 0.5]

pos = nx.spring_layout(Gw, seed=7)  # positions for all nodes - seed for reproducibility
# nodes
nx.draw_networkx_nodes(Gw, pos, node_size=700)

# edges
nx.draw_networkx_edges(Gw, pos, edgelist=elarge, width=6)
nx.draw_networkx_edges(
    Gw, pos, edgelist=esmall, width=6, alpha=0.5, edge_color="b", style="dashed"
)

# node labels
nx.draw_networkx_labels(Gw, pos, font_size=20, font_family="sans-serif")
# edge weight labels
edge_labels = nx.get_edge_attributes(Gw, "weight")
nx.draw_networkx_edge_labels(Gw, pos, edge_labels)

ax = plt.gca()
ax.margins(0.08)
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph with weighted edges
G1 = nx.Graph()
edge_list_with_weights = [('Apple', 'Microsoft', 0.7),
                          ('Apple', 'Google', 0.9),
                          ('Apple', 'Amazon', 0.4),
                          ('Apple', 'Facebook', 0.6),
                          ('Microsoft', 'Google', 0.8),
                          ('Microsoft', 'Netflix', 0.5),
                          ('Google', 'Amazon', 0.3),
                          ('Amazon', 'Facebook', 0.7),
                          ('Facebook', 'Netflix', 0.6),
                          ('Netflix', 'Spotify', 0.2),
                          ('Spotify', 'Uber', 0.1)]

G1.add_weighted_edges_from(edge_list_with_weights)

# Create layout and draw the nodes
pos = nx.spring_layout(G1)
nx.draw(G1, pos, with_labels=True, node_size=2000, node_color="lightblue", font_size=12)

# Draw edges with styles based on weight
nx.draw_networkx_edges(G1, pos, width=[3 if data['weight'] > 0.5 else 1 for u, v, data in G1.edges(data=True)],
                       style=['solid' if data['weight'] > 0.5 else 'dotted' for u, v, data in G1.edges(data=True)])

# Add weights as labels on the edges
edge_labels = nx.get_edge_attributes(G1, 'weight')
nx.draw_networkx_edge_labels(G1, pos, edge_labels={key: f'{val:.2f}' for key, val in edge_labels.items()})

plt.title("Graph with Weighted Edges")
plt.show()
```
