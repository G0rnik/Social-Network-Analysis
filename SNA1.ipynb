```python
import matplotlib as plt
edges = [
   ("nn", "ss", 5), ("nn", "mm", 4), ("nn", "kk", 5), ("nn", "zz", 1), ("nn", "yy", 1),
   ("kk", "ss", 1), ("kk", "rr", 4), ("kk", "mm", 5), ("kk", "zz", 1), ("kk", "yy", 1),
   ("ss", "mm", 3), ("ss", "zz", 3), ("ss", "yy", 3), ("kk", "pp", 1), ("kk", "oo", 1),
   ("rr", "kk", 4), ("rr", "zz", 1), ("rr", "yy", 1), ("rr", "pp", 3), ("rr", "oo", 4),
]

G = nx.Graph()


# Add weighted edges
for src, dst, weight in edges:
   G.add_edge(src, dst, weight=weight)


# Compute inverse weights to make strong edges shorter and weak edges longer
inverse_weights = {edge: 1 / data["weight"] for edge, data in G.edges.items()}


# Compute layout using inverse weights
pos = nx.spring_layout(G, weight="weight", scale=2, k=2.5, seed=42)


# Draw the graph

nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=2000, font_size=10)


# Draw edge labels
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)


# Show plot





# Calculate node strengths
node_strength = {node: sum(data['weight'] for _, _, data in G.edges(node, data=True)) for node in G.nodes()}


# Sort node strengths from highest to lowest
sorted_strengths = dict(sorted(node_strength.items(), key=lambda item: item[1], reverse=True))


# Print sorted strengths
print(f"Node Strengths (Descending): {sorted_strengths}")
```

```python
Gw = nx.Graph()

Gw.add_edge("a", "b", weight=0.6)
Gw.add_edge("a", "c", weight=0.2)
Gw.add_edge("c", "d", weight=0.1)
Gw.add_edge("c", "e", weight=0.7)
Gw.add_edge("c", "f", weight=0.9)
Gw.add_edge("a", "d", weight=0.3)

elarge = [(u, v) for (u, v, d) in Gw.edges(data=True) if d["weight"] > 0.5]
esmall = [(u, v) for (u, v, d) in Gw.edges(data=True) if d["weight"] <= 0.5]

pos = nx.spring_layout(Gw, seed=7)  # positions for all nodes - seed for reproducibility
# nodes
nx.draw_networkx_nodes(Gw, pos, node_size=700)

# edges
nx.draw_networkx_edges(Gw, pos, edgelist=elarge, width=6)
nx.draw_networkx_edges(
    Gw, pos, edgelist=esmall, width=6, alpha=0.5, edge_color="b", style="dashed"
)

# node labels
nx.draw_networkx_labels(Gw, pos, font_size=20, font_family="sans-serif")
# edge weight labels
edge_labels = nx.get_edge_attributes(Gw, "weight")
nx.draw_networkx_edge_labels(Gw, pos, edge_labels)



```

```python
import matplotlib.pyplot as plt


# Create the graph
G = nx.Graph()

# Add edges from the DataFrame
G.add_edges_from(df[['Source', 'Target']].values)


# Generate the layout for the graph
plt.figure(figsize=(30, 16))
pos = nx.spring_layout(G, seed=42)  # Layout for nodes

# Draw the graph with the color map
nx.draw(G, pos, with_labels=True, edge_color="gray", node_size=2000, font_size=10)

# Show the plot
plt.show()
```

```python
# Create the graph
G = nx.Graph()

# Add edges from the DataFrame (assuming 'Source' and 'Target' columns in df)
G.add_edges_from(df[['Source', 'Target']].values)

# Define the color map based on the node names
color_map = []
for node in G.nodes():
    if node.startswith('Customer'):
        color_map.append('blue')  # Color for customers
    elif node.startswith('Influencer'):
        color_map.append('green')  # Color for influencers
    else:
        color_map.append('orange')  # Color for brands or others

# Generate the layout for the graph
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, seed=42)  # Layout for nodes

# Draw the graph with the color map
nx.draw(G, pos, with_labels=True, node_color=color_map, edge_color="gray", node_size=2000, font_size=10)

# Show the plot
plt.show()
```

```python
num_nodes = G.number_of_nodes()
print(f"Number of Nods: {num_nodes}")
```

```python
num_edges = G.number_of_edges()
print(f"Number of Edge: {num_edges}")
```

```python
num_unique_edges = len(set(G.edges()))
print(f"Number of Unique Edges: {num_unique_edges}")
```

```python
# Ensure the graph is connected before calculating diameter
if nx.is_connected(G):
    diameter = nx.diameter(G)
    print(f"Diameter of the Graph: {diameter}")
else:
    print("The graph is not connected. Diameter cannot be computed.")
```

```python
# Calculate Node Size (degree of each node)
node_sizes = {node: G.degree(node) for node in G.nodes()}

# Print Node Sizes in column format
print("Node Sizes (Degree of each node):")
for node, degree in node_sizes.items():
    print(f"{node}: {degree}")
```

```python
# Print Node Sizes in column format
print("Node Sizes (Degree of each node):")
for node, degree in node_sizes.items():
    print(f"{node}: {degree}")

# Extract nodes with degree 5 or more
nodes_with_size_5_or_more = [node for node, degree in node_sizes.items() if degree >= 5]

print("Nodes with degree 5 or more:")
print(nodes_with_size_5_or_more)




# Create a subgraph with nodes that have degree >= 5
subgraph = G.subgraph(nodes_with_size_5_or_more)

# Generate layout for the subgraph
plt.figure(figsize=(10, 6))
pos_subgraph = nx.spring_layout(subgraph, seed=42)  # Layout for nodes

# Draw the subgraph with the same node size and edge color
nx.draw(subgraph, pos_subgraph, with_labels=True, node_color="lightblue", edge_color="gray", node_size=2000, font_size=10)

# Show the plot for the subgraph
plt.show()
```
